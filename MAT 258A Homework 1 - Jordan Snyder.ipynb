```julia
using GLPKMathProgInterface
using Distributions
using PyPlot
using Convex
```

```julia
#define penalty functions
function l1(vec)
    return sum([abs(v) for v in vec])
end
function l2(vec)
    return sum([v^2 for v in vec])
end
function dzlin(a,vec)
    return sum(max(zeros(size(vec)), [abs(v)-a for v in vec]))
end
function logbarrier(a,vec)
    if any([abs(v)>a for v in vec])
        return Inf
    end
    return sum([-log(1-(u/a)^2)*a^2 for u in vec])
end
```
